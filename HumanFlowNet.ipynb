```python
import re
import nltk
import random
import string
import numpy as np
import textstat
import torch
import transformers
import matplotlib.pyplot as plt
from nltk.corpus import wordnet
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
from PyPDF2 import PdfReader, PdfWriter
from reportlab.pdfgen import canvas


nltk.download("punkt")
nltk.download("wordnet")
nltk.download("vader_lexicon")


paraphrase_model = transformers.pipeline("text2text-generation", model="ramsrigouthamg/t5_paraphraser")
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")



def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    texts = []
    for page in reader.pages:
        extracted = page.extract_text()
        if extracted:
            texts.append(extracted)
    return " ".join(texts).strip()

def save_text_to_pdf(text, pdf_path):
    c = canvas.Canvas(pdf_path)
    c.setFont("Helvetica", 12)
    y_position = 800
    for line in text.split("\n"):
        c.drawString(50, y_position, line)
        y_position -= 20
        if y_position < 50:
            c.showPage()
            c.setFont("Helvetica", 12)
            y_position = 800
    c.save()


def get_readability(text):
    return textstat.flesch_reading_ease(text)

def get_lexical_density(text):
    words = nltk.word_tokenize(text)
    unique_words = set(words)
    return len(unique_words) / len(words) if words else 0

def get_punctuation_density(text):
    punctuations = sum(1 for char in text if char in string.punctuation)
    return punctuations / len(text) if text else 0

def compute_tfidf_similarity(text1, text2):
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([text1, text2])
    return cosine_similarity(vectors[0], vectors[1])[0][0]

def compute_embedding_similarity(text1, text2):
    emb1 = embedding_model.encode(text1, convert_to_tensor=True)
    emb2 = embedding_model.encode(text2, convert_to_tensor=True)
    return torch.nn.functional.cosine_similarity(emb1, emb2, dim=0).item()

def replace_with_synonyms(text):
    words = nltk.word_tokenize(text)
    new_words = []
    for word in words:
        synonyms = wordnet.synsets(word)
        if synonyms:
            new_word = synonyms[0].lemmas()[0].name().replace("_", " ")
            new_words.append(new_word)
        else:
            new_words.append(word)
    return " ".join(new_words)

def paraphrase_text(text):
    response = paraphrase_model(text, truncation=True, max_length=512, num_return_sequences=1, do_sample=True, top_k=50, top_p=0.95)
    if isinstance(response, list) and len(response) > 0:
        return response[0].get("generated_text", text)
    return text

def chunk_text_by_words(text, max_words=350):
    words = nltk.word_tokenize(text)
    if len(words) <= max_words:
        return [text]
    chunks = []
    current_chunk = []
    for word in words:
        current_chunk.append(word)
        if len(current_chunk) >= max_words:
            chunks.append(" ".join(current_chunk))
            current_chunk = []
    if current_chunk:
        chunks.append(" ".join(current_chunk))
    return chunks

def modify_text(ai_text, human_text):
    similarity = compute_embedding_similarity(ai_text, human_text)
    chunks = chunk_text_by_words(ai_text, max_words=350)
    modified_chunks = []
    if similarity > 0.85:
        for chunk in chunks:
            modified_chunks.append(replace_with_synonyms(chunk))
    elif similarity > 0.6:
        for chunk in chunks:
            first_pass = paraphrase_text(f"paraphrase: {chunk}")
            second_pass = paraphrase_text(first_pass)
            modified_chunks.append(second_pass)
    else:
        for chunk in chunks:
            paraphrased = paraphrase_text(f"paraphrase: {chunk}")
            modified_chunks.append(replace_with_synonyms(paraphrased))
    modified_text = "\n\n".join(modified_chunks)
    return modified_text, similarity



def get_emotional_intensity(text):
    sia = SentimentIntensityAnalyzer()
    sentences = sent_tokenize(text)
    compound_scores = [sia.polarity_scores(sentence)['compound'] for sentence in sentences]
    return np.mean([abs(score) for score in compound_scores]) if compound_scores else 0

def get_sentence_length_std(text):
    sentences = sent_tokenize(text)
    lengths = [len(nltk.word_tokenize(sentence)) for sentence in sentences]
    return np.std(lengths) if len(lengths) > 1 else 0

def compute_repetition_ratio(text):
    words = nltk.word_tokenize(text.lower())
    bigrams = list(nltk.bigrams(words))
    if not bigrams:
        return 0
    bigram_counts = {}
    for bg in bigrams:
        bigram_counts[bg] = bigram_counts.get(bg, 0) + 1
    repeated = sum(1 for count in bigram_counts.values() if count > 1)
    return repeated / len(bigrams)

def compute_coherence(text):
    sentences = sent_tokenize(text)
    if len(sentences) < 2:
        return 0
    embeddings = embedding_model.encode(sentences, convert_to_tensor=True)
    similarities = []
    for i in range(len(sentences)-1):
        sim = torch.nn.functional.cosine_similarity(embeddings[i], embeddings[i+1], dim=0)
        similarities.append(sim.item())
    return np.mean(similarities)

def extract_features(text):
    features = [
        get_readability(text),
        get_lexical_density(text),
        get_punctuation_density(text),
        get_emotional_intensity(text),
        get_sentence_length_std(text),
        compute_repetition_ratio(text),
        compute_coherence(text)
    ]
    return features



def visualize_features(ai_text, modified_text, human_text):
    features = {
        "Readability": [get_readability(ai_text), get_readability(modified_text), get_readability(human_text)],
        "Lexical Density": [get_lexical_density(ai_text), get_lexical_density(modified_text), get_lexical_density(human_text)],
        "Punctuation Density": [get_punctuation_density(ai_text), get_punctuation_density(modified_text), get_punctuation_density(human_text)]
    }
    labels = ["AI-Generated", "Modified", "Human-Written"]
    plt.figure(figsize=(12, 5))
    for i, (feat, vals) in enumerate(features.items()):
        plt.subplot(1, 3, i+1)
        plt.bar(labels, vals, color=['red', 'blue', 'green'])
        plt.title(feat)
        plt.ylabel("Score")
        plt.xticks(rotation=15)
    plt.suptitle("Feature Comparison: AI vs. Modified vs. Human Text")
    plt.tight_layout()
    plt.show()

def visualize_with_standard(modified_text, baseline_text):
    metrics = {
        "Readability": [get_readability(modified_text), get_readability(baseline_text)],
        "Lexical Density": [get_lexical_density(modified_text), get_lexical_density(baseline_text)],
        "Punctuation Density": [get_punctuation_density(modified_text), get_punctuation_density(baseline_text)],
        "Emotional Intensity": [get_emotional_intensity(modified_text), get_emotional_intensity(baseline_text)],
        "Sentence Length Std": [get_sentence_length_std(modified_text), get_sentence_length_std(baseline_text)],
        "Repetition Ratio": [compute_repetition_ratio(modified_text), compute_repetition_ratio(baseline_text)],
        "Coherence": [compute_coherence(modified_text), compute_coherence(baseline_text)]
    }
    labels = list(metrics.keys())
    modified_vals = [vals[0] for vals in metrics.values()]
    baseline_vals = [vals[1] for vals in metrics.values()]

    angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
    modified_vals += modified_vals[:1]
    baseline_vals += baseline_vals[:1]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    ax.plot(angles, modified_vals, color='blue', linewidth=2, label='Modified Text')
    ax.fill(angles, modified_vals, color='blue', alpha=0.25)
    ax.plot(angles, baseline_vals, color='green', linewidth=2, label='Baseline Standard')
    ax.fill(angles, baseline_vals, color='green', alpha=0.25)
    ax.set_thetagrids(np.degrees(angles[:-1]), labels)
    plt.title("Radar Chart Comparison with Baseline Standard")
    plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
    plt.show()

def radar_chart(features, title="Feature Radar Chart"):
    labels = ["Readability", "Lexical Density", "Punctuation Density", "Emotional Intensity",
              "Sentence Length Std", "Repetition Ratio", "Coherence"]
    angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
    features += features[:1]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    ax.plot(angles, features, linewidth=2, linestyle='solid')
    ax.fill(angles, features, alpha=0.25)
    ax.set_thetagrids(np.degrees(angles[:-1]), labels)
    plt.title(title)
    plt.show()



def plot_prediction_probabilities(rf_model, svm_model, lr_model, human_text, modified_text):
    human_features = extract_features(human_text)
    modified_features = extract_features(modified_text)

    human_prob_rf = rf_model.predict_proba([human_features])[0][1]
    human_prob_svm = svm_model.predict_proba([human_features])[0][1]
    human_prob_lr = lr_model.predict_proba([human_features])[0][1]

    modified_prob_rf = rf_model.predict_proba([modified_features])[0][1]
    modified_prob_svm = svm_model.predict_proba([modified_features])[0][1]
    modified_prob_lr = lr_model.predict_proba([modified_features])[0][1]

    text_types = ['Human Text', 'Modified AI Text']
    rf_probs = [human_prob_rf, modified_prob_rf]
    svm_probs = [human_prob_svm, modified_prob_svm]
    lr_probs = [human_prob_lr, modified_prob_lr]

    plt.figure(figsize=(8,6))
    plt.plot(text_types, rf_probs, marker='o', label='Random Forest')
    plt.plot(text_types, svm_probs, marker='o', label='SVM')
    plt.plot(text_types, lr_probs, marker='o', label='Logistic Regression')
    plt.title("Classifier Prediction Confidence\n(Probability of Being Human)")
    plt.xlabel("Text Type")
    plt.ylabel("Predicted Probability")
    plt.ylim(0, 1)
    plt.legend()
    plt.grid(True)
    plt.show()

def compute_text_flow(text, num_segments=20, feature_func=get_readability):
    words = nltk.word_tokenize(text)
    segment_length = len(words) // num_segments if len(words) >= num_segments else len(words)
    flow_values = []
    for i in range(num_segments):
        segment_words = words[i*segment_length:(i+1)*segment_length]
        if not segment_words:
            break
        segment_text = " ".join(segment_words)
        flow_values.append(feature_func(segment_text))
    return flow_values

def plot_text_flow(human_text, modified_text, num_segments=20, feature_func=get_readability):
    human_flow = compute_text_flow(human_text, num_segments, feature_func)
    modified_flow = compute_text_flow(modified_text, num_segments, feature_func)
    x = list(range(1, len(human_flow)+1))
    plt.figure(figsize=(10,6))
    plt.plot(x, human_flow, marker='o', label='Human Text Flow')
    plt.plot(x, modified_flow, marker='o', label='Modified AI Text Flow')
    plt.title("Text Flow Comparison (Readability Over Segments)")
    plt.xlabel("Segment Number")
    plt.ylabel("Readability Score")
    plt.legend()
    plt.grid(True)
    plt.show()



def train_ai_detector(ai_texts, human_texts):
    features = []
    labels = []
    for text in ai_texts:
        features.append(extract_features(text))
        labels.append(0)
    for text in human_texts:
        features.append(extract_features(text))
        labels.append(1)
    rf_model = RandomForestClassifier(n_estimators=100)
    svm_model = SVC(kernel='linear', probability=True)
    lr_model = LogisticRegression(max_iter=1000)
    rf_model.fit(features, labels)
    svm_model.fit(features, labels)
    lr_model.fit(features, labels)
    return rf_model, svm_model, lr_model



ai_text = extract_text_from_pdf("ai_text.pdf")
human_text = extract_text_from_pdf("In_human_Text.pdf")


modified_text, similarity_score = modify_text(ai_text, human_text)

print("Modified Text (before saving to PDF):")
print(modified_text)
save_text_to_pdf(modified_text, "modified_text.pdf")
print("Modified Text (after saving to PDF):")
print(modified_text)
print("\nEmbedding Similarity Score between AI and Human Text:", similarity_score)

visualize_features(ai_text, modified_text, human_text)


ai_dataset_samples = [
    extract_text_from_pdf("ai_sample_1.pdf"),
    extract_text_from_pdf("ai_sample_2.pdf")
]
human_dataset_samples = [
    extract_text_from_pdf("human_sample_1.pdf"),
    extract_text_from_pdf("human_sample_2.pdf")
]

baseline_text_standard = extract_text_from_pdf("In_human_Text.pdf")


visualize_with_standard(modified_text, baseline_text_standard)


modified_features = extract_features(modified_text)
radar_chart(modified_features, title="Modified Text Full Feature Set")


all_ai_samples = ai_dataset_samples + [ai_text]
all_human_samples = human_dataset_samples + [human_text]
rf_model, svm_model, lr_model = train_ai_detector(all_ai_samples, all_human_samples)


test_features = [extract_features(modified_text)]
rf_prediction = rf_model.predict(test_features)[0]
svm_prediction = svm_model.predict(test_features)[0]
lr_prediction = lr_model.predict(test_features)[0]

print("\nPrediction (1 = AI-generated, 0 = Human-written):")
print("Random Forest:", rf_prediction)
print("SVM:", svm_prediction)
print("Logistic Regression:", lr_prediction)


plot_prediction_probabilities(rf_model, svm_model, lr_model, human_text, modified_text)
plot_text_flow(human_text, modified_text, num_segments=20, feature_func=get_readability)
```
